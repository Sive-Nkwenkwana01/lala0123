```python
import MySQLdb
import matplotlib.pyplot as plt
import pandas as pd
import pymysql
import sqlalchemy
import numpy as np

conn=sqlalchemy.create_engine("mysql+pymysql://root:myheart01@localhost/sakila")

df=pd.read_sql_table("sales",conn)
df
# Data values for creating a bar graph
x = df['PRODUCT NAME']
y = df['SALES AMOUNT']

# Creating a histogram

plt.xticks(rotation=45)
plt.grid(True)

plt.bar(x,y)

plt.xlabel('Products')

plt.ylabel('Sales Amount')

plt.show()
```
